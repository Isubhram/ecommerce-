```python
import time
import requests
import bs4
from bs4 import BeautifulSoup
import urllib
import urllib.request
import urllib.parse
import io
import csv





response = requests.get("https://www.netmeds.com/medicine/manufacturers")
response = response.text
data = bs4.BeautifulSoup(response, 'html.parser')
read1 = data.select("body > div.page-wrapper > div.main-navigation > div > ul > li.m-pro > ul > li:nth-child(1) > a")
print(len(read1))         
```

```python
com_list = []
for i in range(len(read1)):
    x = read1[i].find_all("li")
    for j in range(len(x)):
        y = x[j].find_all("a")
        for k in range(len(y)):
            z = y[k].get("href")
            com_list.append(z)
            print(z)
print(com_list)
```

```python
all_url = []       
for m in range(len(com_list)):
    response = ''
    while response == '':
        try:
            response = requests.get(com_list[m])
            break
        except:
            print("Connection refused by the server..")
            print("Let me sleep for 5 seconds")
            print("ZZzzzz...")
            print("Was a nice sleep, now let me continue...")
            continue
    response = response.text
    data = bs4.BeautifulSoup(response, 'html.parser')
    read2 = data.select(".panel-body")
    for i in range(len(read2)):
        x = read2[i].find_all("li")
        for j in range(len(x)):
            y = x[j].find_all("b")
            for k in range(len(y)):
                z = y[k].get("href")
                all_url.append(z)
print(all_url)

```
